# JOIN In PANDAS

SQL Joins--Joins can be performed with concat,merge and join operations in Pandas

Create a Dataframe

```python
import pandas as pd

raw_data={
    'subject_id':['1','2','3','4','5'],
    'first_name':['Alex','John','Amy','Allen','Nick'],
    'last_name':['Patel','Ali','Dsouza','Aoni','Atiches']}

df_a=pd.DataFrame(raw_data,columns=['subject_id','first_name','last_name'])
df_a
```

Create a second dataframe

```python
raw_data={
    'subject_id':['4','5','6','7','8'],
    'first_name':['Billy','Brain','Alle','Emy','Jack'],
    'last_name':['Lax','Sac','Nac','VAX','Bob']
    
}
df_b=pd.DataFrame(raw_data,columns=['subject_id','first_name','last_name'])
df_b
```

UNION  SELECt * from df1 UNION ALL Select * from df2

Concat the two dataframes along rows,by default axis=0 that is rows

```python
df_new=pd.concat([df_a,df_b],ignore_index=True)
df_new
```

Create a third dataframe

```python
raw_data={
      'subject_id':['1','2','3','4','5','6','7','8'],
       'test_id':[51,25,14,65,45,41,24,58]
}
df_c=pd.DataFrame(raw_data,columns=['subject_id','test_id'])
df_c
```

INNER JOIN
Select * from df1 inner join df2 on df1.key=df2.key;

```python
pd.merge(df_new,df_c)
```

```python
#Merge two dataframes along the subject_id value
#Inside on parameter give the column name which have common data
pd.merge(df_new,df_c,on='subject_id')
```

```python
#Merge with inner join
#Inner join produces only the set of records that match in both Table A and Table B

pd.merge(df_a,df_b, on='subject_id' ,how='inner')
```

```python
#Merge with outer join

#Full outer join produces the set of all records in Table A and Table B

#If there is no match,the missing side will contain null.
pd.merge(df_a,df_b,on='subject_id',how='outer')
```
